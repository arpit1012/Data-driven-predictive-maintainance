```python
#function for preparing training data and forming a RUL column with information about the remaining
```

```python
def prepare_train_data(data, factor = 0):
    df = data.copy()
    fd_RUL = df.groupby('unit_number')['time_in_cycles'].max().reset_index()
    fd_RUL = pd.DataFrame(fd_RUL)
    fd_RUL.columns = ['unit_number','max']
    df = df.merge(fd_RUL, on=['unit_number'], how='left')
    df['RUL'] = df['max'] - df['time_in_cycles']
    df.drop(columns=['max'],inplace = True)

    return df[df['time_in_cycles'] > factor]
```

```python
train_001 = prepare_train_data(train)
```

```python
def score_func(y_true,y_pred,label):
    lst = [
          round(mean_absolute_error(y_true,y_pred),2),
          round(mean_squared_error(y_true,y_pred),2)**0.5,
          round(r2_score(y_true,y_pred),2)]
    
    
    print(f' mean absolute error {lst[0]}',' label =',label)
    print(f' root mean squared error {lst[1]}',' label =',label)
    print(f' R2 score {lst[2]}',' label =',label)
    return [lst[0], round(lst[1],2), lst[2]*100]
```

```python
# We remove the properties that weakly correlate with the
# RUL target: setting_1, setting_2, s6, unit_number, as
# well as one of the features that are highly correlated
# with each other (s9 and s14 have a correlation
# coefficient of 0.96, remove s14)
```

```python
unit_number = pd.DataFrame(train_001["unit_number"])
train_001 = train_001.drop(columns = ['unit_number','setting_1','setting_2','s6','s14'])
train_001.head()
```

```python
#Linear regression
X_train=train_001
y_train=X_train.pop('RUL')
```

```python
drop_labels=['unit_number','setting_1','setting_2','s6','s14','setting_3','s1','s5','s10','s16','s18','s19']

X_test = test.groupby('unit_number').last().reset_index().drop(drop_labels, axis=1)
```

```python
X_test
```

```python
min_max_scaler = MinMaxScaler()



# create and fit model

lm = LinearRegression()
lm.fit(X_train, y_train)



# predict and evaluate
y_hat_train = lm.predict(X_train)
score_func(y_train,y_hat_train,'train')
```

```python


y_hat_test = lm.predict(X_test)
score_func(rul[0],y_hat_test,'test')
```

```python
def plot_sensor(sensor_name):
    plt.figure(figsize=(13,5))
    for i in train['unit_number'].unique():
        if (i % 10 == 0):  # only plot every 10th unit_nr
            plt.plot('RUL', sensor_name, 
                     data=train_001[train_001
                                    ['unit_number']==i])
    plt.xlim(250, 0)  # reverse the x-axis so RUL counts down to zero
    plt.xticks(np.arange(0, 275, 25))
    plt.ylabel(sensor_name)
    plt.xlabel('Remaining Use fulLife')
    plt.show()
```

```python
plot_sensor('s12')
```

```python

## Clipped RUL 
y_train_clipped = y_train.clip(upper=125)

# create and fit model
lm = LinearRegression()
lm.fit(X_train, y_train_clipped)

# predict and evaluate
y_hat_train = lm.predict(X_train)
score_func(y_train_clipped, y_hat_train, 'train')

y_hat_test = lm.predict(X_test)
score_func(rul[0], y_hat_test,'test')
```

```python
from sklearn.svm import SVR
# Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# SVM regression + clipped RUL + scaled
svr = SVR(kernel='linear')
svr.fit(X_train_scaled, y_train_clipped)

# predict and evaluate
y_hat_train = svr.predict(X_train_scaled)
score_func(y_train_clipped, y_hat_train, 'train')

y_hat_test = svr.predict(X_test_scaled)
score_func(rul[0], y_hat_test,'test')
```
